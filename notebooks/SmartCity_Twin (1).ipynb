🌧 Flood Risk Analysis System

Problem Statement

Heavy rainfall can lead to waterlogging and flooding in urban areas, affecting traffic flow, emergency response, and public safety. This module identifies flood-prone sectors of Chandigarh and provides smart recommendations to reduce flood risk.

```python
flood_risk = np.array([
    15,   # Sector 1
    25,   # Sector 14
    55,   # Sector 15
    30,   # Sector 17
    82,   # Sector 22
    45,   # Sector 25
    38,   # Sector 26
    65,   # Sector 32
    40,   # Sector 34
    35,   # Sector 35
    78,   # Sector 43
    20    # Sector 47
])
print(flood_risk)
```

```python
print("="*60)
print("FLOOD RISK DATA")
print("="*60)

for i in range(len(sector_names)):
    print(f"{sector_names[i]} ---> {flood_risk[i]}")
```

```python
print("="*60)
print("FLOOD RISK STATUS")
print("="*60)

for i in range(len(sector_names)):

    if flood_risk[i] <= 30:
        status = "🟢 Low"

    elif flood_risk[i] <= 60:
        status = "🟡 Moderate"

    else:
        status = "🔴 High"

    print(f"{sector_names[i]} ---> {flood_risk[i]} ---> {status}")
```

```python
print("="*60)
print("CHANDIGARH FLOOD DASHBOARD")
print("="*60)

print(f"Average Flood Risk      : {np.mean(flood_risk):.2f}")
print(f"Highest Flood Risk      : {np.max(flood_risk)}")
print(f"Highest Risk Sector     : {sector_names[np.argmax(flood_risk)]}")
print(f"Lowest Flood Risk       : {np.min(flood_risk)}")
print(f"Safest Sector           : {sector_names[np.argmin(flood_risk)]}")
```

```python
print("="*60)
print("SMART FLOOD RECOMMENDATIONS")
print("="*60)

for i in range(len(sector_names)):

    if flood_risk[i] >= 80:
        recommendation = "🚨 Immediate Drainage Improvement Required"

    elif flood_risk[i] >= 60:
        recommendation = "⚠️ Install Additional Water Pumps"

    elif flood_risk[i] >= 40:
        recommendation = "🌧 Regular Monitoring Required"

    else:
        recommendation = "✅ Flood Risk Under Control"

    print(f"{sector_names[i]} ---> {recommendation}")
```

```python
## 3.5 Flood Risk Visualization

import matplotlib.pyplot as plt
import numpy as np

# Calculate average flood risk
average_flood = np.mean(flood_risk)

# Assign colors based on flood risk
colors = []

for value in flood_risk:
    if value <= 30:
        colors.append("green")
    elif value <= 60:
        colors.append("gold")
    else:
        colors.append("red")

# Create Figure
plt.figure(figsize=(14,6))

# Create Bar Chart
bars = plt.bar(
    sector_names,
    flood_risk,
    color=colors,
    edgecolor="black",
    linewidth=1.5
)

# Average Flood Risk Line
plt.axhline(
    y=average_flood,
    color="blue",
    linestyle="--",
    linewidth=2,
    label=f"Average Flood Risk ({average_flood:.2f})"
)

# Show values above each bar
for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 2,
        f"{int(height)}",
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

# Highlight Highest Flood Risk Sector
highest_index = np.argmax(flood_risk)
bars[highest_index].set_color("crimson")

# Graph Title
plt.title(
    "Chandigarh Smart City Flood Risk Dashboard",
    fontsize=18,
    fontweight="bold"
)

# Axis Labels
plt.xlabel(
    "Chandigarh Sectors",
    fontsize=13,
    fontweight="bold"
)

plt.ylabel(
    "Flood Risk Level",
    fontsize=13,
    fontweight="bold"
)

# Rotate Sector Names
plt.xticks(rotation=45)

# Y-axis Limit
plt.ylim(0, 100)

# Grid
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Legend
plt.legend()

# Adjust Layout
plt.tight_layout()
plt.savefig("flood.png", dpi=300, bbox_inches="tight")
# Display Graph
plt.show()
```

```python
print("="*60)
print("FLOOD RISK ANALYSIS SUMMARY")
print("="*60)

print(f"Total Sectors           : {len(sector_names)}")
print(f"Average Flood Risk      : {np.mean(flood_risk):.2f}")
print(f"Highest Flood Risk      : {np.max(flood_risk)}")
print(f"Highest Risk Sector     : {sector_names[np.argmax(flood_risk)]}")
print(f"Lowest Flood Risk       : {np.min(flood_risk)}")
print(f"Safest Sector           : {sector_names[np.argmin(flood_risk)]}")
print(f"Total Flood Risk Score  : {np.sum(flood_risk)}")
```

```python
print("="*60)
print("SMART FLOOD INSIGHTS")
print("="*60)

high = np.sum(flood_risk > 60)
medium = np.sum((flood_risk > 30) & (flood_risk <= 60))
low = np.sum(flood_risk <= 30)

print(f"🟢Low Risk Sectors      : {low}")
print(f"🟡Moderate Risk Sectors : {medium}")
print(f"🔴High Risk Sectors     : {high}")

print("\nHighest Risk Sector :", sector_names[np.argmax(flood_risk)])
print("Safest Sector :", sector_names[np.argmin(flood_risk)])

if np.max(flood_risk )>= 80 :
    print("\nRecommendation :🚨 Immediate drainage improvement required in", sector_names[np.argmax(flood_risk)])
if np.max(flood_risk )> 50 :
    print("Recommendation : 🌧 Upgrade city drainage infrastructure.")
else :
    print("Recommendation :✅ Flood risk is currently under control.")
```

```python
## 4.3 Traffic Comparison (Before vs After Rain)

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(sector_names))
width = 0.35
plt.figure(figsize=(15,6))
bars1 = plt.bar(
    x - width/2,
    traffic,
    width,
    label="Before Rain",
    color="skyblue",
    edgecolor="black"
)
bars2 = plt.bar(
    x + width/2,
    updated_traffic,
    width,
    label="After Rain",
    color="tomato",
    edgecolor="black"
)
# Values above Before Rain bars
for bar in bars1:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 1,
        f"{int(height)}",
        ha="center",
        fontsize=9
    )
# Values above After Rain bars
for bar in bars2:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 1,
        f"{int(height)}",
        ha="center",
        fontsize=9
    )
plt.xticks(x, sector_names, rotation=45)
plt.title(
    "Traffic Comparison Before and After Heavy Rain",
    fontsize=18,
    fontweight="bold"
)
plt.xlabel(
    "Chandigarh Sectors",
    fontsize=13,
    fontweight="bold"
)
plt.ylabel(
    "Traffic Level",
    fontsize=13,
    fontweight="bold"
)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.savefig("Traffic Before and After Rain.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
## 4.4 Rainfall Insights

print("="*60)
print("RAINFALL IMPACT ANALYSIS")
print("="*60)

increase = updated_traffic - traffic
print(f"Average Traffic Before Rain : {np.mean(traffic):.2f} ")
print(f"Average Traffic After Rain : {np.mean(updated_traffic):.2f}")
print(f"\nTotal Increase in Traffic : {np.sum(increase)}")
print(f"Average Increase per Sector : {np.mean(increase):.2f}")
print(f"\nHighest Increase : {np.max(increase)}")
highest_sector = np.argmax(increase)
print("Highest Increase Sector :", sector_names[highest_sector])
```

```python
print("="*60)
print("SMART RAINFALL INSIGHTS")
print("="*60)

for i in range(len(sector_names)):

    if increase[i] >= 8:
        print(f"{sector_names[i]} → 🚨 Severe traffic increase due to rainfall.")
    elif increase[i] >= 5:
        print(f"{sector_names[i]} → ⚠️ Moderate traffic increase.")
    else:
        print(f"{sector_names[i]} → ✅ Minimal rainfall impact.")

print("\nFinal Recommendation")

if np.mean(increase) >= 5:

    print("• Improve city drainage infrastructure.")
    print("• Deploy additional traffic police during heavy rainfall.")
    print("• Display live diversion routes on smart traffic boards.")
    print("• Increase emergency response teams in flood-prone sectors.")

else:
    print("Current rainfall impact is manageable.")
```

🧠 MODULE 8 : SMART DECISION ENGINE

🎯 Problem Statement

A smart city generates data from multiple systems such as traffic, parking, flood monitoring, WiFi coverage, EV charging stations, and emergency services. Instead of analyzing each system independently, the Smart Decision Engine combines all these parameters to identify critical sectors and generate intelligent recommendations for city administrators.

```python
## 8.1 Decision Score Calculation
#Formula - 

Higher Traffic
        +
Higher Parking Occupancy
        +
Higher Flood Risk
        +
Higher EV Demand
        +
Longer Emergency Response
        -
Good WiFi
    ↓
Higher Score
    ↓
More attention needed.
```

```python
decision_score = (
    (0.30 * traffic) +
    (0.20 * parking) +
    (0.25 * flood_risk) +
    (0.15 * ev) +
    (0.10 * response_time) -
    (0.10 * wifi)
)
decision_score = np.round(decision_score, 2)
print(decision_score)
```

## Why Weighted Decision Score?

A smart city contains multiple systems, but not every factor has the same impact on overall city operations.
This Decision Engine assigns different importance (weights) to each parameter based on its influence on urban infrastructure.

- 🚗 Traffic (30%) – Major impact on mobility and congestion.
- 🌧 Flood Risk (25%) – Can severely disrupt transportation and public safety.
- 🅿 Parking (20%) – Indicates infrastructure demand.
- ⚡ EV Charging (15%) – Reflects sustainable transportation needs.
- 🚑 Emergency Response (10%) – Measures service efficiency.
- 📡 WiFi (−10%) – Better connectivity improves city operations, reducing overall risk.

The weighted score provides a more realistic assessment of sector priority compared to treating all factors equally.

```python
print("="*60)
print("SMART CITY RISK CLASSIFICATION")
print("="*60)

for i in range(len(sector_names)):

    if decision_score[i] >= 60:
        status = "🔴 Critical"

    elif decision_score[i] >= 45:
        status = "🟡 Moderate"

    else:
        status = "🟢 Safe"

    print(f"{sector_names[i]} ---> Score = {decision_score[i]} ---> {status}")
```

```python
## 8.3 Smart Decision Dashboard

print("="*60)
print("SMART CITY DECISION DASHBOARD")
print("="*60)

print(f"Average Decision Score : {np.mean(decision_score): .2f}")
print(f"Highest Risk Sector    : {sector_names[np.argmax(decision_score)]}")
print(f"Highest Score          : {np.max(decision_score)}")
print(f"Lowest Risk Score      : {sector_names[np.argmin(decision_score)]}")
print(f"Lowest Score           : {np.mean(decision_score)}")
```

```python
## 8.4 Priority Ranking

print("="*60)
print("SMART CITY PRIORITY RANKING")
print("="*60)

# Sort sectors according to decision score
priority = np.argsort(decision_score)[::-1]

for rank, index in enumerate(priority, start=1):
    print(
        f"Rank {rank:>2} | "
        f"{sector_names[index]:<10} | "
        f"Score = {decision_score[index]:>6.2f}"
    )
```

```python
print("="*60)
print("SMART CITY RECOMMENDATIONS")
print("="*60)

for i in range(len(sector_names)):

    print(f"\n📍 {sector_names[i]}")
    if traffic[i] >= 80:
        print("🚗 Reduce traffic congestion.")
        
    if parking[i] >= 80:
        print("🅿 Expand parking facilities.")
        
    if flood_risk[i] >= 70:
        print("🌧 Improve drainage infrastructure.")
        
    if wifi[i] <= 60:
        print("📡 Install additional WiFi towers.")

    if ev[i] >= 80:
        print("⚡ Increase EV charging stations.")

    if response_time[i] >= 12:
        print("🚑 Deploy additional emergency vehicles.")

    # If no issues found
    if (
        traffic[i] < 80 and
        parking[i] < 80 and
        flood_risk[i] < 70 and
        wifi[i] > 60 and
        ev[i] < 80 and
        response_time[i] < 12
    ):
        print("✅ No immediate action required.")
```

```python
## 8.6 Decision Engine Insights
print("="*60)
print("SMART DECISION ENGINE INSIGHTS")
print("="*60)

critical = np.sum(decision_score >= 60)
moderate = np.sum((decision_score >= 45) & (decision_score < 60))
safe = np.sum(decision_score < 45)

print(f"🔴 Critical Sectors : {critical}")
print(f"🟡 Moderate Sectors : {moderate}")
print(f"🟢 Safe Sectors     : {safe}")

print("\nHighest Priority Sector :", sector_names[np.argmax(decision_score)])
print("Lowest Priority Sector  :", sector_names[np.argmin(decision_score)])

print("\nOverall Recommendation")

if critical >= 3:
    print("🚨 Immediate intervention is required in multiple sectors.")

elif critical >= 1:
    print("⚠️ Focus on high-priority sectors before expanding city infrastructure.")

else:
    print("✅ City conditions are stable.")
```

```python
## 🧠 Smart Decision Engine Dashboard
import matplotlib.pyplot as plt
import numpy as np

# Sort sectors according to decision score
priority = np.argsort(decision_score)[::-1]

sorted_scores = decision_score[priority]
sorted_sectors = np.array(sector_names)[priority]

# Create colors
colors = ["cornflowerblue"] * len(sorted_scores)
colors[0] = "crimson"          # Highest priority
colors[-1] = "mediumseagreen"  # Lowest priority

plt.figure(figsize=(13,7), facecolor="#1E1E1E")

bars = plt.barh(
    sorted_sectors,
    sorted_scores,
    color=colors,
    edgecolor="white",
    linewidth=1.5
)

plt.gca().set_facecolor("#1E1E1E")

# Value Labels
for bar in bars:

    plt.text(
        bar.get_width()+1,
        bar.get_y()+bar.get_height()/2,
        f"{bar.get_width():.1f}",
        va="center",
        color="white",
        fontsize=11,
        fontweight="bold"
    )

# Average Line
avg = np.mean(decision_score)

plt.axvline(
    avg,
    color="yellow",
    linestyle="--",
    linewidth=2,
    label=f"Average Score ({avg:.1f})"
)

plt.title(
    "🧠 Smart Decision Engine Dashboard",
    fontsize=20,
    color="white",
    fontweight="bold"
)

plt.xlabel(
    "Decision Score",
    fontsize=13,
    color="white"
)

plt.ylabel(
    "Sectors",
    fontsize=13,
    color="white"
)

plt.xticks(color="white")
plt.yticks(color="white")

plt.grid(axis="x", alpha=0.3)

plt.legend()

plt.tight_layout()

plt.savefig("decision_engine.png", dpi=300)

plt.show()
```
